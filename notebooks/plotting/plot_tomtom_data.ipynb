```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import datetime as dt
```

```python
gardiner_tomtom_df = pd.read_csv("../csv/gardiner_flow_data.csv")
front_yonge_tomtom_df = pd.read_csv("../csv/front_yonge_flow_data.csv")
```

```python
gardiner_tomtom_df["collectedAt"] = pd.to_datetime(gardiner_tomtom_df["collectedAt"])
front_yonge_tomtom_df["collectedAt"] = pd.to_datetime(front_yonge_tomtom_df["collectedAt"])
```

```python
ax = gardiner_tomtom_df.plot(x="collectedAt", y=["currentSpeed", "currentTravelTime"])
ax.legend(["Current Speed (km/h)", "Current Travel Time (s)"])
ax.set_xlabel("")
plt.savefig("../img/gardiner_current_spd_travel_time_5_min.eps", format="eps")
```

```python
ax = gardiner_tomtom_df.plot(x="collectedAt", y=["freeFlowSpeed", "freeFlowTravelTime"])
ax.legend(["Free Flow Speed (km/h)", "Free Flow Travel Time (s)"])
ax.set_xlabel("")
plt.savefig("../img/gardiner_free_spd_travel_time_5_min.eps", format="eps")
```

```python
ax = front_yonge_tomtom_df.plot(x="collectedAt", y=["currentSpeed", "currentTravelTime"])
ax.legend(["Current Speed (km/h)", "Current Travel Time (s)"])
ax.set_xlabel("")
plt.savefig("../img/front_yonge_current_spd_travel_time_5_min.eps", format="eps")
```

```python
ax = front_yonge_tomtom_df.plot(x="collectedAt", y=["freeFlowSpeed", "freeFlowTravelTime"])
ax.legend(["Free Flow Speed (km/h)", "Free Flow Travel Time (s)"])
ax.set_xlabel("")
plt.savefig("../img/front_yonge_free_spd_travel_time_5_min.eps", format="eps")
```
