```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
veh_can_data = pd.read_csv("../../raw_data/canada_vehicle_registrations.csv")
veh_can_data = veh_can_data.drop(columns=["GEO", "DGUID", "UOM", "UOM_ID", "SCALAR_FACTOR", "SCALAR_ID", "VECTOR", "COORDINATE", "STATUS", "SYMBOL", "TERMINATED", "DECIMALS"])
veh_can_data = veh_can_data[veh_can_data["Type of vehicle"] == "Total, road motor vehicle registrations"]
veh_can_data = veh_can_data.drop(columns=["Type of vehicle"])

veh_ont_data = pd.read_csv("../../raw_data/ontario_vehicle_registrations.csv")
veh_ont_data = veh_ont_data.drop(columns=["GEO", "DGUID", "UOM", "UOM_ID", "SCALAR_FACTOR", "SCALAR_ID", "VECTOR", "COORDINATE", "STATUS", "SYMBOL", "TERMINATED", "DECIMALS"])
veh_ont_data = veh_ont_data[veh_ont_data["Type of vehicle"] == "Total, road motor vehicle registrations"]
veh_ont_data = veh_ont_data.drop(columns=["Type of vehicle"])

pop_can_data = pd.read_csv("../../raw_data/canada_population.csv")
pop_can_data = pop_can_data.drop(columns=["GEO", "DGUID", "Sex", "UOM", "UOM_ID", "SCALAR_FACTOR", "SCALAR_ID", "VECTOR", "COORDINATE", "STATUS", "SYMBOL", "TERMINATED", "DECIMALS"])
pop_can_data = pop_can_data[pop_can_data["Age group"] == "All ages"]

pop_ont_data = pd.read_csv("../../raw_data/ontario_population.csv")
pop_ont_data = pop_ont_data.drop(columns=["GEO", "DGUID", "Sex", "UOM", "UOM_ID", "SCALAR_FACTOR", "SCALAR_ID", "VECTOR", "COORDINATE", "STATUS", "SYMBOL", "TERMINATED", "DECIMALS"])
pop_ont_data = pop_ont_data[pop_ont_data["Age group"] == "All ages"]
```

```python
veh_can_reg_x = veh_can_data["REF_DATE"]
veh_can_reg_x = np.array(veh_can_reg_x).reshape(-1,1)
veh_can_reg_y = veh_can_data["VALUE"]
veh_can_reg_model = LinearRegression()
veh_can_reg_model.fit(veh_can_reg_x, veh_can_reg_y)
veh_can_r_sq = veh_can_reg_model.score(veh_can_reg_x, veh_can_reg_y)
veh_can_intercept = veh_can_reg_model.intercept_
veh_can_slope = veh_can_reg_model.coef_
print(f"Vehicle (Canada) R^2: {veh_can_r_sq}")
print(f"Vehicle (Canada) Intercept: {veh_can_intercept}")
print(f"Vehicle (Canada) Slope: {veh_can_slope}")

veh_ont_reg_x = veh_ont_data["REF_DATE"]
veh_ont_reg_x = np.array(veh_ont_reg_x).reshape(-1,1)
veh_ont_reg_y = veh_ont_data["VALUE"]
veh_ont_reg_model = LinearRegression()
veh_ont_reg_model.fit(veh_ont_reg_x, veh_ont_reg_y)
veh_ont_r_sq = veh_ont_reg_model.score(veh_ont_reg_x, veh_ont_reg_y)
veh_ont_intercept = veh_ont_reg_model.intercept_
veh_ont_slope = veh_ont_reg_model.coef_
print(f"Vehicle (Ontario) R^2: {veh_ont_r_sq}")
print(f"Vehicle (Ontario) Intercept: {veh_ont_intercept}")
print(f"Vehicle (Ontario) Slope: {veh_ont_slope}")

pop_can_reg_x = pop_can_data["REF_DATE"]
pop_can_reg_x = np.array(pop_can_reg_x).reshape(-1,1)
pop_can_reg_y = pop_can_data["VALUE"]
pop_can_reg_model = LinearRegression()
pop_can_reg_model.fit(pop_can_reg_x, pop_can_reg_y)
pop_can_r_sq = pop_can_reg_model.score(pop_can_reg_x, pop_can_reg_y)
pop_can_intercept = pop_can_reg_model.intercept_
pop_can_slope = pop_can_reg_model.coef_
print(f"Population (Canada) R^2: {pop_can_r_sq}")
print(f"Population (Canada) Intercept: {pop_can_intercept}")
print(f"Population (Canada) Slope: {pop_can_slope}")

pop_ont_reg_x = pop_ont_data["REF_DATE"]
pop_ont_reg_x = np.array(pop_ont_reg_x).reshape(-1,1)
pop_ont_reg_y = pop_ont_data["VALUE"]
pop_ont_reg_model = LinearRegression()
pop_ont_reg_model.fit(pop_ont_reg_x, pop_ont_reg_y)
pop_ont_r_sq = pop_ont_reg_model.score(pop_ont_reg_x, pop_ont_reg_y)
pop_ont_intercept = pop_ont_reg_model.intercept_
pop_ont_slope = pop_ont_reg_model.coef_
print(f"Population (Ontario) R^2: {pop_ont_r_sq}")
print(f"Population (Ontario) Intercept: {pop_ont_intercept}")
print(f"Population (Ontario) Slope: {pop_ont_slope}")
```

```python
veh_can_pred_2025 = veh_can_reg_model.predict(np.array([2025]).reshape(-1,1))
veh_can_pred_2050 = veh_can_reg_model.predict(np.array([2050]).reshape(-1,1))
print(f"Canada Predicted vehicle registrations (2025): {veh_can_pred_2025}")
print(f"Canada Predicted vehicle registrations (2050): {veh_can_pred_2050}")

veh_ont_pred_2025 = veh_ont_reg_model.predict(np.array([2025]).reshape(-1,1))
veh_ont_pred_2050 = veh_ont_reg_model.predict(np.array([2050]).reshape(-1,1))
print(f"Ontario Predicted vehicle registrations (2025): {veh_ont_pred_2025}")
print(f"Ontario Predicted vehicle registrations (2050): {veh_ont_pred_2050}")

pop_can_pred_2025 = pop_can_reg_model.predict(np.array([2025]).reshape(-1,1))
pop_can_pred_2050 = pop_can_reg_model.predict(np.array([2050]).reshape(-1,1))
print(f"Canada Predicted population (2025): {pop_can_pred_2025}")
print(f"Canada Predicted population (2050): {pop_can_pred_2050}")

pop_ont_pred_2025 = pop_ont_reg_model.predict(np.array([2025]).reshape(-1,1))
pop_ont_pred_2050 = pop_ont_reg_model.predict(np.array([2050]).reshape(-1,1))
print(f"Ontario Predicted population (2025): {pop_ont_pred_2025}")
print(f"Ontario Predicted population (2050): {pop_ont_pred_2050}")
```

```python
sns.regplot(x=pop_can_reg_x, y=pop_can_reg_y, label="Population")
# plt.plot(pop_can_reg_x, pop_can_reg_model.predict(pop_can_reg_x), color="gray", label="Pop. Regression")
sns.regplot(x=veh_can_reg_x, y=veh_can_reg_y, label="Veh. Registration")
# plt.plot(veh_can_reg_x, veh_can_reg_model.predict(veh_can_reg_x), color="k", label="Veh. Regression")
plt.xlabel("Year")
plt.ylabel("Vehicles Registered/Country Population")
plt.xticks(np.arange(veh_can_reg_x[0], veh_can_reg_x[len(veh_can_reg_x)-1], step=3))
plt.legend(loc="upper left")
# plt.tick_params(axis="y", which="both", left=False, right=False, labelleft=False)
plt.savefig("../../img/canada_registration_population.eps", format="eps", bbox_inches="tight")
plt.savefig("../../img/canada_registration_population.pdf", bbox_inches="tight")
```
