```python
front_sample = front_df.sample(1)
front_sample = front_df[(front_df["collectedAt"].dt.day.isin(front_sample["collectedAt"].dt.day)) & (
    front_df["collectedAt"].dt.month.isin(front_sample["collectedAt"].dt.month))]
front_sample = front_sample.sort_values(by="collectedAt")
front_sample = front_sample.groupby("collectedAt").mean().reset_index()
```

```python
front_thunderstorm = front_df[front_df["weather"].isin([202, 211])]
front_rain = front_df[front_df["weather"].isin([500, 501, 502, 520])]
front_snow = front_df[front_df["weather"].isin([600, 601])]
front_mist = front_df[front_df["weather"].isin([701, 721, 741])]
front_clear = front_df[front_df["weather"].isin([800])]
front_fog = front_df[front_df["weather"].isin([801, 801, 802, 803])]
```

```python
front_thunderstorm["rain1h10e-3"] = front_thunderstorm["rain1h"].apply(lambda x: x * 1000)
front_rain["rain1h10e-3"] = front_rain["rain1h"].apply(lambda x: x * 1000)
front_snow["snow1h10e-3"] = front_snow["snow1h"].apply(lambda x: x * 1000)
```
