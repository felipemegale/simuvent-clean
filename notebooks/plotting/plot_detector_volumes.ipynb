```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# bay has data from July 01-07
# Spadina has data from August 11-17
bay_loc = pd.read_csv("../csv/gardiner_bay_count.csv")
spadina_loc = pd.read_csv("../csv/gardiner_spadina_count.csv")
```

```python
bay_loc = bay_loc.drop(columns=["detector_id", "number_of_lanes", "primary_road",
                       "detector_loc", "latitude", "longitude", "distance_to_union_km"])
spadina_loc = spadina_loc.drop(columns=["detector_id", "number_of_lanes",
                               "primary_road", "detector_loc", "latitude", "longitude", "distance_to_union_km"])
```

```python
bay_eastbound = bay_loc[bay_loc["direction"] == "E"]
bay_westbound = bay_loc[bay_loc["direction"] == "W"]

spadina_eastbound = spadina_loc[spadina_loc["direction"] == "E"]
spadina_westbound = spadina_loc[spadina_loc["direction"] == "W"]
```

```python
fig, ax = plt.subplots()

bay_eastbound_plot = ax.plot(
    bay_eastbound["datetime_bin"], bay_eastbound["volume_15min"], label="Eastbound traffic vol.")
bay_westbound_plot = ax.plot(
    bay_westbound["datetime_bin"], bay_westbound["volume_15min"], label="Westbound traffic vol.")
ax.legend()
original_xticks = ax.get_xticks()

fig.tight_layout()

plt.xticks([original_xticks[0], original_xticks[-1]])

plt.savefig("../img/gardiner_bay_traffic_volumes.eps", format="eps")
```

```python
fig, ax = plt.subplots()

spadina_eastbound_plot = ax.plot(
    spadina_eastbound["datetime_bin"], spadina_eastbound["volume_15min"], label="Eastbound traffic vol.")
spadina_westbound_plot = ax.plot(
    spadina_westbound["datetime_bin"], spadina_westbound["volume_15min"], label="Westbound traffic vol.")
ax.legend()
original_xticks = ax.get_xticks()

fig.tight_layout()

plt.xticks([original_xticks[0], original_xticks[-1]])

plt.savefig("../img/gardiner_spadina_traffic_volumes.eps", format="eps")
```
