```python
import pandas as pd
from geopy import distance
YONGE_LOCS_MIDDLE_POINT = (43.693963, -79.395695)
```

If loc lat > loc middle point lat, loc is north of point (+)
If loc lat < loc middle point lat, loc is south of point (-)

```python
df = pd.read_csv("../../csv/v2/030_yonge_count_processed.csv.zip")
df.head(3)
```

```python
len(df)
```

```python
df.sort_values(by="time_end", inplace=True)
df
```

# Expand datetime_bin

```python
df["time_end"] = pd.to_datetime(df["time_end"],utc=True)
df["year"] = df["time_end"].dt.year
df["month"] = df["time_end"].dt.month_name()
df["day"] = df["time_end"].dt.day
df["day_of_week"] = df["time_end"].dt.day_name()
df.head(3)
```

# Filter df to have only Yonge St related data

```python
df = df[df["location"].str.contains("YONGE ST")]
df
```

# Understand categorical features

## Detector location

```python
len(df["location"].drop_duplicates())
```

```python
df["location"].value_counts()
```

# Understanding numerical features

## _id
- Just a numerical field telling which reading this is
    - can be dropped

```python
df.drop("_id",inplace=True,axis=1)
df.head()
```

## Latitude and longitude
- These attributes will be useful
    - By determining the distance from a middle point
    - Create two new cols telling whether the point is north or south from the middle point

```python
def calculate_distance(coord):
    return distance.distance(YONGE_LOCS_MIDDLE_POINT,coord).km

df["dist_from_mid_point"] = df.apply(lambda x: calculate_distance((x["lat"],x["lng"])),axis=1)
df.head(3)
```

```python
df["is_north_from_mid_point"] = df.apply(lambda x: 1 if x["lat"] > YONGE_LOCS_MIDDLE_POINT[0] else 0,axis=1)
df.head(3)
```

- Now I can drop lat and lng because they were already used to calculate the distance from the middle point

```python
df = df.drop(["lat","lng"],axis=1)
df.head(3)
```

- Can also drop distance_from_point

```python
df.drop("distance_from_point",inplace=True,axis=1)
df.head(3)
```

## Date and time columns
- Drop year beacause data is seasonal
- Drop end_day_of_week because will onehot encode day of week name

```python
df = df.drop(["year","end_day_of_week"],axis=1)
df.head(3)
```

# Rename a few columns

```python
df.rename(columns={
    "time_end":"time",
    "end_hour":"hour",
    "end_minute":"minute",
    "end_day_of_week":"day_of_week",
},inplace=True)

df.head(3)
```

```python
min_vol = df["total_count"].min()
max_vol = df["total_count"].max()
mean_vol = df["total_count"].mean()
median_vol = df["total_count"].median()
variance = df["total_count"].var()
std_dev = df["total_count"].std()

print(
    f"Min volume:    {min_vol}\n"
    f"Max volume:    {max_vol}\n"
    f"Mean volume:   {mean_vol}\n"
    f"Median volume: {median_vol}\n"
    f"Variance:     {variance}\n"
    f"Std dev:      {std_dev}"
)
```
