```python
import pandas as pd
```

```python
yonge_df = pd.read_csv("./data/202305_yonge_flow_data.csv.zip")
weather_df = pd.read_csv("./data/202305_weather_data.csv.zip")
```

```python
yonge_df.collectedAt = pd.to_datetime(yonge_df.collectedAt,utc=True)
yonge_df.collectedAt = yonge_df.collectedAt.dt.floor(freq="t")
weather_df.collectedAt = pd.to_datetime(weather_df.collectedAt,utc=True)
weather_df.collectedAt = weather_df.collectedAt.dt.floor(freq="t")
```

```python
df = yonge_df.merge(weather_df,on="collectedAt",how="inner")
df
```

```python
df["Month"] = df.collectedAt.dt.month_name()
df["Weekday"] = df.collectedAt.dt.day_name()
df["Day"] = df.collectedAt.dt.day
df["Hour"] = df.collectedAt.dt.hour
df["Minute"] = df.collectedAt.dt.minute
df.drop(["collectedAt", "windDegree"],inplace=True,axis=1)
df
```

```python
categorical_attrs = ["Month", "Weekday"]

transformed_df = pd.get_dummies(df, columns=categorical_attrs, prefix="", prefix_sep="")
transformed_df = transformed_df.astype("float32")
transformed_df
```

```python
transformed_df.columns
```

```python
transformed_df.rename({
    'visibility':"Visibility",
    'temperature':"Temperature",
    'FeelsLike':"Feels Like",
    'humidity':"Humidity",
    'windSpeed':"Wind Speed",
    'cloudsAll':"Clouds All",
    'rain1h':"Rain (1h)",
    'snow1h':"Snow (1h)",
    'cloudy': "Cloudy",
    'clear': "Clear",
    'atmosphere': "Atmosphere",
    'snowy': "Snowy",
    'rainy':"Rainy"
}, axis=1, inplace=True)
transformed_df
```

```python
transformed_df.to_csv("./data/202305_yonge_flow_weather.csv.zip", index=False, compression="zip")
```
